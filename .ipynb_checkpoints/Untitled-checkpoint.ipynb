```python
import requests
import pandas as pd
import matplotlib.pyplot as plt

# --- Координаты (Шымкент, можешь изменить на свой регион) ---
latitude = 42.3
longitude = 69.6

# --- Дата начала и окончания периода ---
start = 20220101
end = 20221231

# --- Параметры: температура воздуха и скорректированные осадки ---
parameters = "T2M,PRECTOTCORR"

# --- Сборка URL ---
url = f"https://power.larc.nasa.gov/api/temporal/daily/point?parameters={parameters}&community=AG&longitude={longitude}&latitude={latitude}&start={start}&end={end}&format=JSON"

# --- Запрос данных ---
response = requests.get(url)
data = response.json()

# --- Проверим, какие данные пришли ---
print("Доступные параметры:", data['properties']['parameter'].keys())

# --- Извлечение и построение таблицы ---
records = data['properties']['parameter']
dates = list(records['T2M'].keys())

df = pd.DataFrame({
    'date': pd.to_datetime(dates),
    'temperature_C': list(records['T2M'].values()),
    'precipitation_mm': list(records['PRECTOTCORR'].values())  # ← здесь обновлено!
})

# --- Сохраняем в CSV (на всякий случай) ---
df.to_csv("weather_kazakhstan.csv", index=False)

# --- Показываем первые строки таблицы ---
df.head()
```

```python
# --- Устанавливаем дату как индекс ---
df.set_index('date', inplace=True)

# --- Температура ---
plt.figure(figsize=(12, 4))
df['temperature_C'].plot(title='Температура воздуха (°C)', color='orange')
plt.ylabel('°C')
plt.grid(True)
plt.show()

# --- Осадки ---
plt.figure(figsize=(12, 4))
df['precipitation_mm'].plot(title='Осадки (мм)', color='blue')
plt.ylabel('мм')
plt.grid(True)
plt.show()
```

```python
# --- Простая логика классификации условий ---
def classify_conditions(temp, rain):
    if rain < 5:
        return "Засуха"
    elif rain > 20:
        return "Переувлажнение"
    elif temp < 5:
        return "Слишком холодно"
    else:
        return "Нормально"

# --- Применим к каждому дню ---
df['condition'] = df.apply(lambda row: classify_conditions(row['temperature_C'], row['precipitation_mm']), axis=1)

# --- Посмотрим результат ---
df[['temperature_C', 'precipitation_mm', 'condition']].head(10)
```

```python
# --- Подсчитаем сколько дней с каждым типом условий ---
df['condition'].value_counts()
```

```python
# --- Покажем график температуры с цветами по условиям ---
colors = {
    "Нормально": "green",
    "Засуха": "red",
    "Переувлажнение": "blue",
    "Слишком холодно": "gray"
}

plt.figure(figsize=(14, 5))
for condition, color in colors.items():
    subset = df[df['condition'] == condition]
    plt.scatter(subset.index, subset['temperature_C'], label=condition, color=color, s=10)

plt.title("Классификация погодных условий по температуре")
plt.xlabel("Дата")
plt.ylabel("Температура (°C)")
plt.grid(True)
plt.legend()
plt.show()
```

```python
# --- Простая логика рекомендаций ---
def give_recommendation(condition):
    if condition == "Нормально":
        return "Подходит для посадки"
    elif condition == "Засуха":
        return "Риск засухи – подождите"
    elif condition == "Переувлажнение":
        return "Переувлажнение – не рекомендуется"
    elif condition == "Слишком холодно":
        return "Слишком холодно – не сажать"
    else:
        return "Нет данных"

# --- Добавим колонку с рекомендациями ---
df['recommendation'] = df['condition'].apply(give_recommendation)

# --- Посмотрим примеры ---
df[['temperature_C', 'precipitation_mm', 'condition', 'recommendation']].head(15)
```

```python
from datetime import datetime

# Отфильтруем нормальные дни
df_sorted = df.reset_index().sort_values('date')
today = datetime(2022, 4, 1)  # ← Можешь поменять на любую дату

# Ищем "подходит для посадки" после сегодняшней даты
next_window = df_sorted[(df_sorted['date'] > today) & 
                        (df_sorted['recommendation'] == 'Подходит для посадки')]

# Покажем ближайшие 7 дней с хорошими условиями
next_window[['date', 'temperature_C', 'precipitation_mm', 'recommendation']].head(7)
```

```python
# Когда начинается и заканчивается благоприятный период
good_days = next_window['date'].tolist()

if good_days:
    start = good_days[0]
    end = good_days[-1]
    print(f"🌱 Следующее подходящее окно посадки: с {start.date()} по {end.date()}")
else:
    print("⚠️ В ближайшие дни нет благоприятных условий для посадки.")
```

```python
app.py
import streamlit as st
import pandas as pd
import matplotlib.pyplot as plt

# Загружаем данные
df = pd.read_csv("weather_kazakhstan.csv", parse_dates=['date'])
df.set_index('date', inplace=True)

# Классификация условий
def classify_conditions(temp, rain):
    if rain < 5:
        return "Засуха"
    elif rain > 20:
        return "Переувлажнение"
    elif temp < 5:
        return "Слишком холодно"
    else:
        return "Нормально"

def give_recommendation(condition):
    if condition == "Нормально":
        return "✅ Подходит для посадки"
    elif condition == "Засуха":
        return "⚠️ Риск засухи – подождите"
    elif condition == "Переувлажнение":
        return "⚠️ Переувлажнение – не рекомендуется"
    elif condition == "Слишком холодно":
        return "❄️ Слишком холодно – не сажать"
    else:
        return "Нет данных"

df['condition'] = df.apply(lambda row: classify_conditions(row['temperature_C'], row['precipitation_mm']), axis=1)
df['recommendation'] = df['condition'].apply(give_recommendation)

# Интерфейс
st.title("🌾 Прогноз для фермера (на основе данных NASA)")
st.write("Выберите диапазон дат:")

date_range = st.date_input("Диапазон", [df.index.min().date(), df.index.max().date()])

if len(date_range) == 2:
    start_date, end_date = date_range
    filtered_df = df[(df.index >= pd.to_datetime(start_date)) & (df.index <= pd.to_datetime(end_date))]

    st.subheader("📈 Температура воздуха")
    st.line_chart(filtered_df['temperature_C'])

    st.subheader("🌧️ Осадки")
    st.line_chart(filtered_df['precipitation_mm'])

    st.subheader("📋 Рекомендации по дням")
    st.dataframe(filtered_df[['temperature_C', 'precipitation_mm', 'condition', 'recommendation']])
```
